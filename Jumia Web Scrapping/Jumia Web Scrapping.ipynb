```python
#General
import pandas as pd
from bs4 import BeautifulSoup as bs
import re

#for offline
import urllib.parse, urllib.request

#for online
import requests
```

# Offline Version

```python
url_jumia ="file:///C:/xampp/htdocs/xampp/websites%20for%20scrapping/Computers%20-%20Buy%20Laptops,%20Desktops%20&%20Accessories%20_%20Online%20_%20Jumia%20Kenya.html"   
#replace above with an oflline jumia link you have

dt_url_jumia = urllib.request.urlopen(url_jumia).read()
url_jumia_dt = bs(dt_url_jumia,'lxml')

#class is a key word , we use class_
url_jumia_data = url_jumia_dt.find_all('div', class_ = 'sku -gallery')

bidhaa_name =[bidhaa.find('span',class_ = 'name').text for bidhaa in url_jumia_data ]
bidhaa_brand = [bidhaa.find('span',class_ = 'brand').text for bidhaa in url_jumia_data]
bidhaa_price = [bidhaa.find('span',class_='price').text for bidhaa in url_jumia_data]
bidhaa_discounts = [bidhaa.find('span', class_ = 'sale-flag-percent') for bidhaa in url_jumia_data]
bidhaa_ratings = [bidhaa.find('div', class_='total-ratings') for bidhaa in url_jumia_data ]


bidhaa_rating = []
bidhaa_discount = []

#cover the aspect of some goods not having a rating
#results in nonetype error
for item in bidhaa_ratings:
    if item!= None:
        item = item.text.strip('()')
        bidhaa_rating.append(item)
    else:
         bidhaa_rating.append(item)

#cover the aspect of some goods not having a discount
#results in nonetype error
for item in bidhaa_discounts:
    if item !=None:
        item = item.text
        bidhaa_discount.append(item)
        
    #adds the nonetype value to the list
    else:
        bidhaa_discount.append(item)
        
    

data_dict_to_csv = {'Name':bidhaa_name,'Brand':bidhaa_brand,'Price':bidhaa_price,'Discount':bidhaa_discount,'Rating':bidhaa_rating}

jumia_csv = pd.DataFrame(data_dict_to_csv, columns=['Name','Brand','Price','Discount','Rating'])

print(jumia_csv)

```

## For Multiple Pages  i.e using category links

```python
def scrap_data(link):
    
    
    url_jumia = requests.get(link)
    url_jumia_dt = bs(url_jumia.content,'lxml')

    #class is a key word , we use class_
    url_jumia_data = url_jumia_dt.find_all('div', class_ = 'sku -gallery')

    bidhaa_name =[bidhaa.find('span',class_ = 'name').text for bidhaa in url_jumia_data ]
    bidhaa_brand = [bidhaa.find('span',class_ = 'brand').text for bidhaa in url_jumia_data]
    bidhaa_price = [bidhaa.find('span',class_='price').text for bidhaa in url_jumia_data]
    bidhaa_discounts = [bidhaa.find('span', class_ = 'sale-flag-percent') for bidhaa in url_jumia_data]
    bidhaa_reviews = [bidhaa.find('div', class_='total-ratings') for bidhaa in url_jumia_data ]

    bidhaa_review = []
    bidhaa_discount = []

    #cover the aspect of some goods not having a review
    #results in nonetype error
    for item in bidhaa_reviews:
        if item!= None:
            item = item.text.strip('()')
            bidhaa_review.append(item)
        else:
             bidhaa_review.append(item)

    #cover the aspect of some goods not having a discount
    #results in nonetype error
    for item in bidhaa_discounts:
        if item !=None:
            item = item.text
            bidhaa_discount.append(item)

        #adds the nonetype value to the list
        else:
            bidhaa_discount.append(item)


            
    #get the name of the excel sheet based on the webpage name
    excel_name = link.split('/')[-2]
    excel_name_to_use = excel_name+'.xlsx'
    
    
    data_dict_to_csv = {'Name':bidhaa_name,'Brand':bidhaa_brand,'Price':bidhaa_price,'Discount':bidhaa_discount,'Reviews':bidhaa_review}
    jumia_csv = pd.DataFrame(data_dict_to_csv, columns=['Name','Brand','Price','Discount','Reviews'])
    #check results in the excel sheets
    jumia_csv.to_excel(excel_name_to_use,index=False)
    
   
 
    


base_url = 'https://www.jumia.co.ke/'

u_r_l = requests.get(base_url)

bs_url = bs(u_r_l.content,'lxml')

bs_url_data = bs_url.find_all('a' , class_ ='itm' , role = 'menuitem')

#print(bs_url_data)

links=[]

for link in bs_url_data:
    link=re.findall('^<a class="itm" href=.([^"]+)*',str(link))
    
    if len(link)==1:
        for lin in link:
            if lin.endswith('/') == True:
                   links.append(link[0])
    

#add the base url to all paths in the main page to turn them to websites
links= [urllib.parse.urljoin(base_url, url) for url in links]


#loop through the links in the list
for link in links:
    scrap_data(link)
```

## For (multiple) all the pages on the base page i.e the pages that appear numbered, 1,2,3,4,...,12,...,19,...,30

```python
#any jumia category link/searched link should work
url = requests.get('https://www.jumia.co.ke/computing/')
url_bs = bs(url.content,'lxml')

all_available_pages = url_bs.find_all('li',class_ ='item')
all_page_links = []

for page in all_available_pages:
    page_links = re.findall('href=.([^"]+)*',str(page))
    all_page_links.append((page_links[0]))

#list of all available pages
all_pages_str = []
all_pages_int = []

for link in all_page_links:
    page_no = re.findall('page.([^ ]+)*',str(link))
    if len(page_no)!= 0 :
        all_pages_str.append(page_no[0])

    
for page_int in all_pages_str:
    if page_int.isdigit():
        if page_int not in all_pages_int:
            #convert the strings to integers so that the max  function can work
            #had to convert because it was not picking 25
            #and also for the arnge function to work
            all_pages_int.append(int(page_int))
            
highest_page = max(all_pages_int)
    
#list to get alll the items in all the pages
total_bidhaa_name =[]
total_bidhaa_brand = []
total_bidhaa_price = []
total_bidhaa_discount = []
total_bidhaa_reviews = []
    
for page in range(1,highest_page):
    #get the url page but add the page value at the end
    page_url  = 'https://www.jumia.co.ke/computing/?page='+str(page)
    
    data = requests.get(page_url)
    data_cont = bs(data.content,'lxml')

    url_jum = data_cont.find_all('div', class_='sku -gallery' ) 
    
    bidhaa_name =[bidhaa.find('span',class_ = 'name').text for bidhaa in url_jum ]
    bidhaa_brand = [bidhaa.find('span',class_ = 'brand').text for bidhaa in url_jum]
    bidhaa_price = [bidhaa.find('span',class_='price').text for bidhaa in url_jum]
    bidhaa_discounts = [bidhaa.find('span', class_ = 'sale-flag-percent') for bidhaa in url_jum]
    bidhaa_reviews = [bidhaa.find('div', class_='total-ratings') for bidhaa in url_jum ]
    
    
    
    bidhaa_review = []
    bidhaa_discount = []

    #cover the aspect of some goods not having a review
    #results in nonetype error
    for item in bidhaa_reviews:
        if item!= None:
            item = item.text.strip('()')
            bidhaa_review.append(item)
        else:
             bidhaa_review.append(item)

    #cover the aspect of some goods not having a discount
    #results in nonetype error
    for item in bidhaa_discounts:
        if item !=None:
            item = item.text
            bidhaa_discount.append(item)

        #adds the nonetype value to the list
        else:
            bidhaa_discount.append(item)
    
    #put all values from seperate pages into one list
    total_bidhaa_name.extend(bidhaa_name)
    total_bidhaa_brand.extend(bidhaa_brand)
    total_bidhaa_price.extend(bidhaa_price)
    total_bidhaa_discount.extend(bidhaa_discount)
    total_bidhaa_reviews.extend(bidhaa_review)

all_the_items_data = {'Name':total_bidhaa_name,'Brand':total_bidhaa_brand,'Price':total_bidhaa_price,
                      'Discount':total_bidhaa_discount,'Reviews':total_bidhaa_reviews}

all_the_items_data_df=pd.DataFrame(all_the_items_data, columns=['Name','Brand','Price','Discount','Reviews'])


#check results in the csv
all_the_items_data_df.to_excel("All Items.xlsx", index=False)

```

## Get data from individual webpages of an item that was in the base page

```python
u_r_l=input("Enter URL: ")#like one with the formart shown below: Has to be a Jumia category/search page
#we shall use the link for computing .....  >>   https://www.jumia.co.ke/computing/

url_jumia = requests.get(u_r_l)
#beautifulsoup object
url_jumia_dt = bs(url_jumia.content,'lxml')

#class is a key word , we use class_
url_jumia_data = url_jumia_dt.find_all('div', class_ = 'sku -gallery')


#get individual item's details
bidhaa_link = [bidhaa.find('a', class_ = 'link') for bidhaa in url_jumia_data]
all_individual_page_links = []
#get all individual item links
for bidhaa in bidhaa_link:
    ac_link = re.findall('^<a class="link" href=.([^"]+)*',str(bidhaa))
    all_individual_page_links.append(ac_link[0]) 

#lists to store all the item variables later on
bidhaa_name=[]
bidhaa_brand=[]
bidhaa_new_price=[]
bidhaa_old_price=[]
bidhaa_discounts=[]
bidhaa_reviews=[]
bidhaa_ratings=[]
bidhaa_actual_ratings=[]   
    

#get the details,details of the item
for link in all_individual_page_links:
    actual_item = requests.get(link)
    actual_item_bs = bs(actual_item.content,'lxml')

    #dat_bs = bs_url.find_all('div', class_='row card _no-g -fh -pas')
    item_name = actual_item_bs.find('h1', class_= '-fs20 -pts -pbxs').text
    item_brand = actual_item_bs.find('a', class_='_more').text

    item_reviews = actual_item_bs.find('a', class_='-plxs -fs14 _more')
    if item_reviews != None:
        #deal with non numeric values that are not 'None'    
        try:
            #get the text with reviews, remove , the brackets and get the first value
            item_reviews = float((((item_reviews.text).strip('()').split()))[0])
        except:
            item_reviews = 0
    else:
        item_reviews = 0


    item_rating = actual_item_bs.find('div', class_='stars _s _al')
    if item_rating != None:
        #get the text , then split, then get the first value which is the rating
            try:
                item_rating = float(((item_rating.text).split())[0])
            except:
                item_rating = 0
    else:
        item_rating = 0

    #get the actual rating and round it up to one decimal point
    actual_item_rating = round((item_rating*item_reviews+5+1)/(item_reviews+2) , 2)

    #old price, new price, discount
    item_new_price = actual_item_bs.find('span', class_ = '-b -ltr -tal -fs24').text
    item_old_price = actual_item_bs.find('span', class_ = '-tal -gy5 -lthr -fs16').text
    item_discount = actual_item_bs.find('span',class_='tag _dsct _dyn -mls').text
    
    
    
    ##adding all individual element data to the gerneral_list
    bidhaa_name.append(item_name)
    bidhaa_brand.append(item_brand)
    bidhaa_old_price.append(item_old_price)
    bidhaa_discounts.append(item_discount)
    bidhaa_new_price.append(item_new_price)
    bidhaa_reviews.append(item_reviews)
    bidhaa_ratings.append(item_rating)
    bidhaa_actual_ratings.append(actual_item_rating)
    
    
    
data_dict_to_csv = {'Name':bidhaa_name,'Brand':bidhaa_brand,'Old Price':bidhaa_old_price,'Discount':bidhaa_discounts,'New Price':bidhaa_new_price
                    ,'Reviews':bidhaa_reviews,'Seen Rating':bidhaa_ratings,'Actual Rating':bidhaa_actual_ratings}

jumia_csv = pd.DataFrame(data_dict_to_csv, columns=['Name','Brand','Old Price','Discount', 'New Price','Reviews',
                                                    'Seen Rating','Actual Rating'])


#sending to an excel sheet
#jumia_csv.to_excel('Get data from individual webpages of an item that was in the base page.xlsx', index=False)


print(jumia_csv)
```
