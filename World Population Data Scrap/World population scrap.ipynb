# WORLDMETERS POPULATION SCRAP

```python
import requests
import urllib.request,urllib.parse
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
```

```python
url= "file:///C:/Users/Aseta/Downloads/to web scrap world meter/Population%20by%20Country%20(2020)%20-%20Worldometer.html"
```

```python
data = urllib.request.urlopen(url).read()
soup = bs(data,'lxml')
soup_data = soup.find_all('table' , class_ = "table table-striped table-bordered dataTable no-footer")
```

```python
for a in soup_data:
    rows = a.find_all('tr')
```

```python
#Getting the table headings of the data
list_of_headings =[]
for row in rows:
    headings = row.find_all('th')
    if headings!=0:
        for heading in headings:
            heading=heading.text
            list_of_headings.append(heading)
            
#list_of_headings
```

```python
len(list_of_headings)

#use the length to know the number of lists to use
```

```python
list_0,list_1,list_2,list_3,list_4,list_5,list_6,list_7,list_8,list_9,list_10,list_11 = [],[],[],[],[],[],[],[],[],[],[],[]
list_of_lists=[list_0,list_1,list_2,list_3,list_4,list_5,list_6,list_7,list_8,list_9,list_10,list_11]
list_of_lists
```

```python
#creating the lists to store the actual data
for row in rows:
    cell_data = row.find_all("td")
    country_data = []
    for data in cell_data:
        country_data.append(data)
    if len(country_data)!=0:
        n=0
        for item in country_data:
            data_to_add = item.text
            list_of_lists[n].append(data_to_add)
            #print(data_to_add,n)
            n=n+1    
    
#print(list_2)        
```

```python
#Put the data in a adataframe and a csv
data_dict = {list_of_headings[0]:list_0, list_of_headings[1]:list_1 ,list_of_headings[2]:list_2 ,list_of_headings[3]:list_3,
             list_of_headings[4]:list_4 ,list_of_headings[5]:list_5 ,list_of_headings[6]:list_6 ,list_of_headings[7]:list_7 ,
             list_of_headings[8]:list_8 ,list_of_headings[9]:list_9 ,list_of_headings[10]:list_10 ,list_of_headings[11]:list_11}

the_dataframe = pd.DataFrame(data_dict, columns= [list_of_headings[0],list_of_headings[1],list_of_headings[2],list_of_headings[3],
                                                   list_of_headings[4],list_of_headings[5],list_of_headings[6],list_of_headings[7],
                                                   list_of_headings[8],list_of_headings[9],list_of_headings[10],list_of_headings[11]])
```

```python
the_dataframe.to_excel("worldmeter_population.xlsx",index=False)
the_dataframe
```
